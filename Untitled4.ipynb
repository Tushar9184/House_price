```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("juhibhojani/house-price")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
import os

df = pd.read_csv(os.path.join(path, "house_prices.csv"))
df.head()
```

```python
df["Floor"].value_counts()
df["floor"]=df["Floor"].str.split(" ").str[0].str.lower()
df["floor"]=df["floor"].replace({
    "lower":-1,
    "ground":0,
    "upper":0,
    "basement":-1,
})
df["floor"]=pd.to_numeric(df["floor"],errors="coerce")
df["floor"]=df.groupby("location")["floor"].transform(lambda x: x.fillna(x.median()))
df["floor"] = df["floor"].fillna(df["floor"].median())
```

```python
df["BHK"]=df["Title"].str.split(" ").str[0]
```

```python

df['Description'] = df['Description'].fillna('').str.lower()


df["is_furnished"] = df["Description"].apply(lambda x: 1 if "furnished" in x else 0)
df["has_parking"] = df["Description"].apply(lambda x: 1 if "parking" in x else 0)
df["is_vastu"] = df["Description"].apply(lambda x: 1 if "vastu" in x else 0)


df["new_cont"] = df["Description"].apply(lambda x: 1 if any(word in x for word in ["new", "under construction", "possession"]) else 0)
```

```python
df= df.drop("Description",axis=1)
```

```python
df.head()
```

```python
df= df.drop(columns=["Title","Plot Area","Dimensions"],axis=1)
```

```python
df["Rup"]=df["Amount(in rupees)"].str.split(" ").str[0]
df["iscr"] = df["Amount(in rupees)"].apply(lambda x: 100 if "Lac"in x else 1)
df["Amount in cr"]= pd.to_numeric(df["Rup"], errors='coerce')/df["iscr"]
```

```python
df= df.drop(columns=["Rup","iscr","Amount(in rupees)"],axis=1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df2= pd.get_dummies(df["location"],dtype=int)
df2.head()
```

```python
df=df.merge(df2,left_index=True,right_index=True)
```

```python
carpet_area_numeric = pd.to_numeric(df["Carpet Area"].str.split(" ").str[0], errors='coerce')
super_area_numeric = pd.to_numeric(df["Super Area"].str.split(" ").str[0], errors='coerce')
df["Total Area"] = carpet_area_numeric.fillna(super_area_numeric)
```

```python
df.drop(columns=["Carpet Area","Super Area","location","Price (in rupees)"],inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Status"].value_counts()
```

```python
df["status"]=df["Status"].apply(lambda x: 1 if x=="Ready to Move" else 0)
```

```python
df["Transaction"].value_counts()
```

```python
df.dropna(subset=['Transaction'], inplace=True)
```

```python
df = pd.get_dummies(df, columns=['Transaction'], drop_first=True, dtype=int)
```

```python
df["Furnishing"].value_counts()
```

```python

df.loc[(df['Furnishing'].isnull()) & (df['is_furnished'] == 1), 'Furnishing'] = 'Furnished'
df['Furnishing'] = df['Furnishing'].fillna('Semi-Furnished')
```

```python

cols={
    "Furnished":3,
    "Semi-Furnished":2,
    "Unfurnished":1
}
df['Furnishing'] = df['Furnishing'].map(cols)
```

```python
df["Bathroom"].value_counts()
```

```python
df["BHK"] = pd.to_numeric(df["BHK"], errors='coerce')
df["BHK"] = df["BHK"].fillna(df["BHK"].median())
```

```python
df["Bathroom"]
```

```python
df["Bathroom"]=pd.to_numeric(df["Bathroom"],errors="coerce")
df["Bathroom"]=df["Bathroom"].apply(lambda x: x if x<6 else 6)
```

```python
df['Bathroom'] = df.groupby('BHK')['Bathroom'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 1))
```

```python
df["Ownership"].value_counts()
```

```python
df_temp = df.copy()
df_temp['Ownership'] = df_temp['Ownership'].fillna('Unknown')
ownership_dummies = pd.get_dummies(df_temp['Ownership'], prefix='Ownership', dtype=int)
correlation_df = ownership_dummies.join(df_temp['Amount in cr'])
print(correlation_df.corr()['Amount in cr'])
```

```python
df["Society"].value_counts()
```

```python
df['is_in_society'] = df['Society'].notnull().astype(int)
```

```python
print(df.isnull().sum())
```

```python
df["Balcony"].value_counts()
```

```python
df["Balcony"]=pd.to_numeric(df["Balcony"],errors="coerce")
df["Balcony"]=df["Balcony"].apply(lambda x: x if x<6 else 6)
```

```python
df["Balcony"]=df.groupby("BHK")["Balcony"].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 1))
```

```python
df["facing"]
```

```python
df['facing'] = df['facing'].fillna('Unknown')
df['facing'] = df['facing'].str.lower().str.replace(' ', '').str.replace('-', '')
df = pd.get_dummies(df, columns=['facing'], drop_first=True, dtype=int)
```

```python
df["overlooking"]
```

```python

df['overlooking'] = df['overlooking'].fillna('Unknown')

df['view_garden'] = df['overlooking'].str.contains('Garden', case=False).astype(int)
df['view_pool'] = df['overlooking'].str.contains('Pool', case=False).astype(int)
df['view_main_road'] = df['overlooking'].str.contains('Main Road', case=False).astype(int)
df['view_park'] = df['overlooking'].str.contains('Park', case=False).astype(int)


df.drop(columns=['overlooking'], inplace=True)
```

```python
df["Car Parking"]
```

```python

def get_parking_score(text):
    if pd.isna(text):
        return 0

    text = str(text).lower()
    if 'covered' in text:
        return 2
    if 'open' in text:
        return 1
    return 0

df['parking_count'] = df['Car Parking'].str.extract(r'(\d+)').fillna(0).astype(int)

df['parking_score'] = df['Car Parking'].apply(get_parking_score)


df.loc[df['parking_count'] > 5, 'parking_count'] = 0


df.drop(columns=['Car Parking'], inplace=True)

print(df[['parking_score', 'parking_count']].value_counts())
```

```python
df.dropna(subset=['Amount in cr'], inplace=True)
```

```python

cols_with_nulls = df.columns[df.isnull().any()].tolist()


if cols_with_nulls:
    print(f"Deleting the following columns due to null values: \n{cols_with_nulls}")


    df.drop(columns=cols_with_nulls, inplace=True)
else:
    print("No columns with null values found. Data is clean!")

print("\nRemaining columns in the dataset:")
print(df.columns.tolist())
```

```python
# 1. Remove extreme price outliers
# Keeping only properties between 5 Lakhs (0.05 Cr) and 15 Crores (15.0 Cr)
df = df[(df['Amount in cr'] >= 0.05) & (df['Amount in cr'] <= 15.0)]

# 2. Remove area outliers (e.g., 1 sqft or 1,000,000 sqft)
df = df[(df['Total Area'] > 150) & (df['Total Area'] < 10000)]

print(f"Rows remaining after filtering: {len(df)}")
```

```python
from sklearn.linear_model import LinearRegression
y=df["Amount in cr"]
X=df.drop(columns=["Amount in cr"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor


model = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)

for col in X_train.columns:
    if X_train[col].dtype == 'object':
        X_train[col] = pd.to_numeric(X_train[col], errors='coerce')
    X_train[col] = X_train[col].fillna(X_train[col].median())

model.fit(X_train, y_train)
```

```python
for col in X_test.columns:
    if X_test[col].dtype == 'object':
        X_test[col] = pd.to_numeric(X_test[col], errors='coerce')
    X_test[col] = X_test[col].fillna(X_test[col].median())
y_predict= model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
